<h2 STYLE= "background-color:#0f3460;text-align:center;color:#e0e0e0;">DRAW SHAPES</h2>

<h2 STYLE= "background-color:#0f3460;text-align:center;color:#e0e0e0;">IMAGE THRESHOLD</h2>

```python
import cv2

photo = cv2.imread('photo.jpg')
gray = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)

threshold_value = 120
max_value = 255

_, binary_img = cv2.threshold(gray, threshold_value, max_value, cv2.THRESH_BINARY_INV)

cv2.imshow('Binary Threshold', binary_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h2 STYLE= "background-color:#0f3460;text-align:center;color:#e0e0e0;">ADAPTIVE THRESHOLD</h2>

```python
import cv2

photo = cv2.imread('photo.jpg')
gray = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)

block_size = 15
constant = 4

adaptive_result = cv2.adaptiveThreshold(
    gray,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    block_size,
    constant
)

cv2.imshow('Adaptive Threshold', adaptive_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h2 STYLE= "background-color:#0f3460;text-align:center;color:#e0e0e0;">EDGE DETECTION</h2>

```python
import cv2

photo = cv2.imread('photo.jpg')

blurred = cv2.GaussianBlur(photo, (5, 5), 0)

lower_threshold = 50
upper_threshold = 150

detected_edges = cv2.Canny(blurred, lower_threshold, upper_threshold)

cv2.imshow('Detected Edges', detected_edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h2 STYLE= "background-color:#0f3460;text-align:center;color:#e0e0e0;">CONTOURS</h2>

```python
import cv2

photo = cv2.imread('photo.jpg')
gray = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)

all_contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print(f'Total contours found: {len(all_contours)}')

output = photo.copy()
cv2.drawContours(output, all_contours, -1, (0, 165, 255), 2)

cv2.imshow('Contours Detected', output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h2 STYLE= "background-color:#0f3460;text-align:center;color:#e0e0e0;">FACE DETECTION</h2>

```python
import cv2

photo = cv2.imread('photo.jpg')
gray = cv2.cvtColor(photo, cv2.COLOR_BGR2GRAY)

classifier_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
face_detector = cv2.CascadeClassifier(classifier_path)

scale = 1.1
min_neighbors = 6

detected_faces = face_detector.detectMultiScale(gray, scaleFactor=scale, minNeighbors=min_neighbors)

print(f'Faces detected: {len(detected_faces)}')

for (x, y, w, h) in detected_faces:
    cv2.rectangle(photo, (x, y), (x + w, y + h), (0, 165, 255), 2)
    cv2.putText(photo, 'Face', (x, y - 8), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 165, 255), 2)

cv2.imshow('Face Detection Result', photo)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
