```python
# Install OpenCV if needed
# !pip install opencv-python

import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("sample.jpg")   # put your image in the same folder

# Convert BGR (OpenCV default) to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
```

```python
cv2.imwrite("output_saved.jpg", img)
print("Image saved as output_saved.jpg")
```

```python
# Resize to half
resized = cv2.resize(img_rgb, (300, 300))

# Crop image
crop = img_rgb[50:250, 50:250]

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(resized)
plt.title("Resized Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("off")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

```python
draw_img = img_rgb.copy()

# Draw line
cv2.line(draw_img, (50, 50), (250, 50), (255, 0, 0), 3)

# Draw rectangle
cv2.rectangle(draw_img, (50, 100), (200, 250), (0, 255, 0), 3)

# Draw circle
cv2.circle(draw_img, (300, 200), 50, (0, 0, 255), 4)

# Put text
cv2.putText(draw_img, "OpenCV Demo", (50, 300),
            cv2.FONT_HERSHEY_SIMPLEX, 1, (255,255,255), 2)

plt.imshow(draw_img)
plt.title("Shapes and Text")
plt.axis("off")
```

```python
# Simple threshold
_, th1 = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

# Adaptive threshold
th2 = cv2.adaptiveThreshold(gray, 255,
                            cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                            cv2.THRESH_BINARY, 11, 2)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(th1, cmap="gray")
plt.title("Simple Threshold")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(th2, cmap="gray")
plt.title("Adaptive Threshold")
plt.axis("off")
```

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis("off")
```

```python
contours, _ = cv2.findContours(th1, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

contour_img = img_rgb.copy()
cv2.drawContours(contour_img, contours, -1, (0,255,0), 2)

plt.imshow(contour_img)
plt.title("Contours Detected")
plt.axis("off")
```

```python
# Load Haar Cascade file (download from OpenCV GitHub if not present)
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

faces = face_cascade.detectMultiScale(gray, 1.3, 5)

face_img = img_rgb.copy()
for (x, y, w, h) in faces:
    cv2.rectangle(face_img, (x, y), (x+w, y+h), (255,0,0), 2)

plt.imshow(face_img)
plt.title("Face Detection")
plt.axis("off")
```
