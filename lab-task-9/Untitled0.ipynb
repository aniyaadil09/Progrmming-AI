STEP 3 — Import Libraries

```python
import nltk
import spacy
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.wsd import lesk
from textblob import TextBlob
```

STEP 4 — Load spaCy Model

```python
nlp = spacy.load("en_core_web_sm")
print("spaCy loaded ")
```

STEP 5 —  Text

```python
text = """
OpenAI announced a new GPT model in March 2026.
Elon Musk visited the AI summit in Berlin last week.
Microsoft invested $2 billion in an AI startup in San Francisco.
Apple and Google are competing to release new AI products this year.
"""
```

STEP 6 — Text Preprocessing

```python
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Normalize
text_norm = text.lower().strip()

# Tokenization
sentences = sent_tokenize(text_norm)
words = word_tokenize(text_norm)

# Remove stopwords & punctuation
words_clean = [w for w in words if w.isalnum() and w not in stop_words]

# Stemming
words_stemmed = [stemmer.stem(w) for w in words_clean]

# Lemmatization
words_lemmatized = [lemmatizer.lemmatize(w) for w in words_clean]

print("Sentences:", sentences)
print("Clean Words:", words_clean)
print("Stemmed:", words_stemmed)
print("Lemmatized:", words_lemmatized)
```

STEP 7 — POS Tagging

```python
pos_tags = nltk.pos_tag(words_lemmatized)
print("POS Tags:")
print(pos_tags)
```

STEP 8 — NER using NLTK

```python
print("NER (NLTK):")

for sentence in sentences:
    tokens = nltk.word_tokenize(sentence)
    tags = nltk.pos_tag(tokens)
    tree = nltk.ne_chunk(tags)

    for subtree in tree:
        if hasattr(subtree, 'label'):
            entity = " ".join([token for token, pos in subtree])
            print(entity, "->", subtree.label())
```

STEP 9 — NER using spaCy

```python
print("NER (spaCy):")

doc = nlp(text)
for ent in doc.ents:
    print(ent.text, "->", ent.label_)
```

STEP 10 — Relationship Extraction

```python
print("Relationship Extraction :")

for sent in sentences:
    print("Sentence:", sent)

    tokens = nltk.word_tokenize(sent)
    tags = nltk.pos_tag(tokens)

    nouns = [word for word, pos in tags if pos.startswith('NN')]
    verbs = [word for word, pos in tags if pos.startswith('VB')]

    print("Nouns:", nouns)
    print("Verbs:", verbs)
```

STEP 11 — Word Sense Disambiguation (WSD)

```python
print("Word Sense Disambiguation:")

words_to_check = ['model', 'summit', 'startup', 'release']

for word in words_to_check:
    sense = lesk(words, word)
    if sense:
        print(word, ":", sense.definition())
    else:
        print(word, ": No sense found")
```
