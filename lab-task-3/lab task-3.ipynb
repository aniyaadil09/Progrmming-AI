```python

jugA_capacity = 4
jugB_capacity = 3
goal = 2
```

 #  DFS function

```python

def dfs(state, visited, path):
    A, B = state

    if A == goal or B == goal:
        path.append((state, "Goal reached"))
        return path

    visited.add(state)
    path.append((state, ""))

    moves = [
        ((jugA_capacity, B), "Fill Jug A"),
        ((A, jugB_capacity), "Fill Jug B"),
        ((0, B), "Empty Jug A"),
        ((A, 0), "Empty Jug B"),
        ((A - min(A, jugB_capacity-B), B + min(A, jugB_capacity-B)), "Pour Jug A -> B"),
        ((A + min(B, jugA_capacity-A), B - min(B, jugA_capacity-A)), "Pour Jug B -> A")
    ]

    for next_state, rule in moves:
        if next_state not in visited:
            path_copy = path.copy()
            path_copy[-1] = (state, rule)
            result = dfs(next_state, visited, path_copy)
            if result:
                return result

    return None
```

# Run DFS

```python

state = (0, 0)

solution = dfs(state, set(), [])


if solution:
    print("Steps to reach the goal:\n")
    step = 1
    for jug, rule in solution:
        if rule != "":
            print(f"Step {step}: {rule} -> Jug A = {jug[0]}, Jug B = {jug[1]}")
            step += 1
else:
    print("No solution found.")
```
